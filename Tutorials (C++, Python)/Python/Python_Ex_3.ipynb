Note that when no value of the variable is passed the function will use the default parameter value

````python
# Recursion

def factorial(n):
  if n > 1 :
    return n*factorial(n-1)
  else:
    return 1

print(factorial(3))
````

Recusion is a powerful tool which uses functions. Let's break down the above code:

factorial(3), will call the factorial(n) with n = 3, hence the statement 3 x factorial(2) will be executed. Now since we do not have value of factorial(2), pyhton will wait for the calculation of factorial(2), hence the factorial(n) function will be called again with n = 2, this will execute 2 x factorial(1). The factorial(n) function will be called again with n = 1, which will return 1. 

Now the substitution of value will be done as shown below:

factorial(2) = 2 x factorial(1) = 2 x 1 = 2

factorial(3) = 3 x factorial(2) = 3 x 2 = 6

### Classes/Objects

You will hear this sentence a lot, "Python is an object oriented language". What does that mean? 

It means that we can make objects in pyhton, an object contains methods and properties. 

````python
# how to define a class

class Person: 
  def __init__(self, name, age): 
    self.name = name  # property of this class
    self.age = age  # another property of this class

  def father(self, father_name): # method of this class
    print(father_name +" is the father of " + self.name) # note the use of pluse in the print function
````

All the properties are defined in the in ```
__init__()
```function

Every funtion inside a class must take `self` as input



`name` and `age` are the variable which will be passed onto the class when we use it. These variables are stored as `self.name` and `self.age` as a part of the object. 

All the methods which are defined in the class can use the stored variable e.g.`self.name`

We can pass extra variables to the methods, e.g. `surname` 

````python
# How to use the class

person1 = Person("Robert", 46) # this would call the __init__() function, which "name" and "age" as the input variable 

# person1 class is now defined, hence the self.name and self.age are stored in this class

# self.name variable can be accessed outside the class by OBJECT_NAME.name i.e. person1.name

print(person1.name) 
print(person1.age)
````

`person1` is also called as _instance_ of the class `Person` 

We never give the value of self when we use it. E.g. `__init__()` is defined with 3 variables: `self`, `name`, `age`. But we only gave 2 variables: `Robert` as `name` and `46` as `age`. `self` is not given to the function

````python
# methods can be accessed with OBJECT_NAME.METHODS_NAME(VARIBALES)

person1.father("John") # Note that self variable is not given, only the surname variable is give a value "John"
````

The above code called the method `father()` of the the instance `person1` of class `Person`. Which executed the print command in it. `self.name` is already stored in this class and `father_name` is given the value `"John"`. Note again the `self` variable is not provided with any value. 

````python
# another instance

person2 = Person("Tina", 19)
print(person2.name)
print(person2.age)
````

We make many instances of a class, just like we can use a function many times.

````python
# change the properties of an instance

person1.name = "Ryan"

print(person1.name)
person1.father("John") 
````

### Inheritance

In general classes can have a lot of properties and function. In order to make a similar class, rather than writing the entire class again from scatch, we could inherit all the properties of the original class (Parent class) to the new class (Child class)

````python
# Inheritance

class Employee(Person): # class Employee is exactly same as class Person
  pass

employee_1 = Employee("Emma", 26)
print(employee_1.name)
````

Class `Employee` is same as class `Person`. Hence when we use it, we need to give the same initializing variables as we did for class `Person`

`pass` command help us in situation where we do not want to write something in the code but have to. 

````python
# example of using pass other than in class

i = 3
if i > 2:
  pass # doesn't do anything

def my_fun():
  pass # doesn't do anything
````

````python
# Add properties and method to the child class

class Employee(Person):
  def __init__(self, var_1, var_2, company): # note that __init__() in child class must require atleast as many variables as in __init__() of the parent class
    super().__init__(var_1, var_2) # importe self.name and self.age from the parent class
    self.company = company # note that the we could use self.company but it is not necessary

  def employee_id(self):
    print(self.name + " works at " + self.company)
````

````python
# using the employee class with added properties and methods
employee_2 = Employee("Georg", 30, "ABC_firm")

employee_2.employee_id()
````
