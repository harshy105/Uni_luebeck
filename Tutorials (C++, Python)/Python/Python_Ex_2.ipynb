## Data Structure

In Python the data structure is done using List, Tuple, Set and Dictionaries. 



### List

````python
a_list = ["Python","C++",3, 3.02, True, 3]
print(a_list)
````

### Tuple

````python
a_tuple = ("Python","C++",3, 3.02, True, 3)
print(a_tuple)
````

### Sets

````python
a_set = {"Python","C++",3, 3.02, True, 3}
print(a_set)
# Note that duplicate are removed in set
````

### Dictionary

````python
a_dict = {"new_lang": "Python",
          "old_lang": "C++",
          "integer_1": 3,
          "flt": 3.02, 
          "boolean": True, 
          "integer_2": 3}
print(a_dict)
# Note the ":" operator used in the dictionary
````

##### List are most commonly used as data structure followed by dictonary and Tuple. Set are not used in common practise. Hence in this exercise we will resrict ourselves to List, Dictionay and Tuple. Let's us first see the major operation that could be performed on these data type. 

## Data structure Operations

### Access the element

````python
print(a_list[1])
print(a_tuple[1])
print(a_dict["old_lang"])
````

##### Note the use of Square brackets [] for accessing the element
Dictionary are useful in the sense that in order to access an element, we don't need to remember the index of it. 

### Changing the element

````python
a_list[1] = "C"
a_dict["old_lang"] = "C"
print(a_list)
print(a_dict)

a_tuple[1] = "C"
````

Tuple are immutable, meaning that they can not be changed

````python
# Change the element in Tuple: convert tuple to list and then update the list and convert this list back to the tuple

b_list = list(a_tuple) # this creates a list same as a_tuple
b_list[1] = "C"
a_tuple = tuple(b_list) # create tuple from the list
print(a_tuple)
````

### Adding a new item

````python
a_list.append("hello_world")
print(a_list)
````

## Loops

### While loop

Let's see a simple example first

````python
itr = 0
while itr < 6:
  print(itr)
  itr = itr + 1
````

In the above example,

```
  print(itr)
  itr = itr + 1
```
 
is exected as again and again as long as the 

```
itr < 6
```
condtion is true, since the itr value increase by 1 in every loop, after 6 iteration, while condition will not be valid




````python
# Try to predict the output of the code and then compare your result

itr = 0
while itr < 6:
  itr = itr + 1
  print(itr)
````

````python
# Try to predict the output of the code and then compare your result

itr = 0
while itr <= 6:
  print(itr)
  itr = itr + 1
````

````python
# Try to predict the output of the code and then compare your result

itr = 0

print("start")

while itr < 6:
  print(itr)
  itr = itr + 1

print("End")
````

````python
# Break statement

itr = 0
while itr < 6:
  print(itr)
  itr = itr + 1
  if itr > 3:
    break
````

In the above code, when the itr = 4, break statement is executed, which makes the program break out of the while loop even if the while condition is true. 

### For loop

For loop iterates over list, tuple, dictionary, set and string

````python
for i in a_list:
  print(i)
````

````python
for i in a_tuple:
  print(i)
````

````python
for i in a_set:
  print(i)
````

````python
for i in a_dict:
  print(i)
````

````python
a_string = " Exam ple"
for i in a_string:
  print(i)
````

````python
# Break statement

for i in a_string:
  print(i)
  if i == "m":  # Note the "==" sign, it is used whenever an EQUALITY CONDITION is tested
    break
````

````python
# Running a loop till certain number; using the range function

for i in range(4): # it stops at 4
  print(i)
````

Note that in the above code the index starts from 0 and not 1. In Python the first element is 0 and not 1. \\

Also range(4) does not contain 4.

````python
# changing the start element

for i in range(2,5): # 2 is the start, 5 is stop
  print(i)
````

````python
# changing the step size

for i in range(3,11,2): # 3 is the start, 11 is the stop, 2 is the step size
  print(i)
````

You might have noticed that number of parameters which can be given to a function in python can vary, e.g. we can give 1, 2 or 3 parameters to range function. \\
Based of the number of parameters that are given, range function uses them in a certain manner. In most of the documentation range() is defined like range(*start, stop, *step). Here the parameter which are marked * are optional. 

````python
# having a reversed iteration through number, set step as a negative number

for i in range(10, 2, -2):
  print(i)
````
