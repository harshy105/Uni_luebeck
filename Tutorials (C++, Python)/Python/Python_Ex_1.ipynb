## Conditional: If-Else

let's seen as example of Conditional 

```python
x = 2
if x > 0:
  print("x is greater than zero")
if x < 0:
  print("x is smaller than zero")
```

The output of the above code is "x is greater than zero". let's break down this code:
- x > 0 : will return a boolen "True" 
- if (True): will instruct that the command in the next line should executed
- x < 0 : will return a boolen "False"
- if (False) : will instruct that the command in the next line should not be executed. Hence we don't see "x is smaller than zero" in the output. 


```python
x = 2
if x > 0:
  print("x is greater than zero")
  print("x > 0")
if x < 0:
  print("x is smaller than zero")
  print("x < 0")
```

Observe that both the command under "if x > 0:" are executed and the command under  "if x < 0:" are not executed

```python
x = 2
if x < 0:
  print("x is smaller than zero")
  print('Hi')
print("x < 0")
```

"print("x is smaller than zero")" and  "print('Hi')" are the part of the if command. Hence in order to write the command under the if statement, one must provide indent at the start of the line.




```python
x = 1

if x > 0:
  print("x > 0")
else:
  print("x < 0")
```
